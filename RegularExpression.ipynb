```python
# 정규표현식
# 문자열에서 특정 패턴에 매칭하는 문자열을 찾아내거나 수정하는 방식

# 적용방식
# 1. "정규표현식 패턴" 객체를 생성
# 2. "메소드"를 통해서 문자열에서 패턴에 매칭되는 문자열을 찾아내거나 수정
```

```python
import re
```

```python
# 1). "정규표현식 패턴" 객체를 생성

reg = re.compile("\d\d\d-\d\d\d-\d\d\d\d")  # \d - 숫자
```

```python
# 2). "메소드"를 통해서 문자열에서 패턴에 매칭되는 문자열을 찾아내거나 수정
# search - 패턴이 일치되는 첫 번쨰 문자열만 리턴한다..

mo = reg.search('My number is 415-555-2323')
mo.group()
```

```python
# 2. Search() - 매칭이 되는 첫번째 문자열..리턴
```

```python
reg = re.compile("Batman|Tina Fey")  # "|" => or
mo = reg.search("Batman and Tina Fey is a movie")
mo.group()
```

```python
reg = re.compile("Bat(wo)?man")  # "?" => 앞에 정의된 패턴이 1번 또는 0번 반복..패턴
```

```python
mo = reg.search("The Adventures of Batman")
mo.group()
```

```python
mo = reg.search("The Adventures of Batwoman")
mo.group()
```

```python
mo = reg.search("The Adventures of Batwowoman")
mo.group()
```

```python
reg = re.compile("Bat(wo)+man")  # "+" => 앞에 정의된 패턴이 1번 이상 등장..패턴
```

```python
mo = reg.search("The Adventures of Batman")
mo.group()
```

```python
mo = reg.search("The Adventures of Batwoman")
mo.group()
```

```python
mo = reg.search("The Adventures of Batwowoman")
mo.group()
```

```python
reg = re.compile("Bat(wo)*man")  # "*" => 앞에 정의된 패턴이 0번 이상 등장..패턴
```

```python
mo = reg.search("The Adventures of Batman")
mo.group()
```

```python
mo = reg.search("The Adventures of Batwoman")
mo.group()
```

```python
mo = reg.search("The Adventures of Batwowowoman")
mo.group()
```

```python
reg = re.compile("Bat(wo){2}man")  # {n} => 앞에 정의된 패턴이 n번 반복..패턴
```

```python
mo = reg.search("The Adventures of Batman")
mo.group()
```

```python
mo = reg.search("The Adventures of Batwoman")
mo.group()
```

```python
mo = reg.search("The Adventures of Batwowoman")
mo.group()
```

```python
reg = re.compile("Bat(wo){2,4}man")  # {n, m} => 앞에 정의된 패턴이 n번부터 m번까지 반복..패턴
```

```python
mo = reg.search("The Adventures of Batman")
mo.group()
```

```python
# 정규표현식 반복패턴

# ? - 0번 또는 1번 반복
# * - 0번 이상 반복
# + - 1번 이상 반복
# {n} - n번 반복
# {n,m} - n번부터 m번까지 반복
```

```python
# 3. findall()
```

```python
reg = re.compile("\d\d\d-\d\d\d-\d\d\d\d")
reg.findall("my number is 214-874-8965 and your number is 287-845-8956")
```

```python
# 정규표현식 패턴

# \d - 숫자
# \s - 빈칸, 탭, 줄바꿈
# \w - 문자, 숫자, 밑줄('_')

# \D - 숫자가 아닌 문자
# \S - 빈칸, 탭, 줄바꿈이 아닌 문자
# \W - 문자, 숫자, 밑줄이 아닌 문자
```

```python
# 사용자가 정의 정규표현식 패턴

# [] - 사용자 정의 정규표현식
# [aeiouAEIOU] - 모음
# [a-zA-Z] - 알파벳
# [0-9] - 숫자
# [a-zA-Z0-9] - 알파벳 혹은 숫자
# [^patter] - patter이 아닌 문자 (not)
```

```python
reg = re.compile('\d+\s\w+')
reg.findall("12 drums, 11 pipers, 10 lords, 0 ladies, 8 maid, 7 swans")
```

```python
reg = re.compile('[aeiouAEIOU]')
reg.findall('RoboCop eats baby food. BABY FOOD.')
```

```python
reg = re.compile('[^aeiouAEIOU]')
reg.findall('RoboCop eats baby food. BABY FOOD.')
```

```python
# ^ -  문자열 시작부분...
# $ - 문자열 끝 부분...
```

```python
reg = re.compile("^Hello")  # 문자열 시작에 나오는 Hello
reg.findall("Hello world")
```

```python
reg = re.compile("^Hello")  # 문자열 시작에 나오는 Hello
reg.findall("Park, Hello")
```

```python
reg = re.compile('\d+$')
reg.findall("Your number55 is 430")
```

```python
# 이메일 추출
text = '저의 이메일은 pine@hanmail.net이고, 담당자 이메일은 fio@gamil.com입니다'

reg = re.compile('[a-zA-Z0-9]+@[a-zA-Z0-9]+.[a-z]+')
reg.findall(text)
```

```python
# 금액만 추출
text = '총 금액은 122330.6원입니다. 세금은 5214.233원입니다'

reg = re.compile(r'\d+.\d+원')
reg.findall(text)
```

```python
# 날짜만 추출
text = "계약날짜는 2015년 2월 15일, 계약 종료일은 2017년 06월 22일입니다"

reg = re.compile(r"\d+년\s\d+월\s\d+일")
reg.findall(text)
```

```python
# url을 추출

text = ''''
저의 홈페이지 주소는 https://www.homepage.com입니다.
당신의 홈페이지 주소는 http://www.yourhomepage.com입니다'''

reg = re.compile(r'http[s]?://w{3}\.[a-zA-Z0-9_-]+\.[a-z]+')
reg.findall(text)
```

```python
# url을  비인식 => url을 "*****"로 바꾸시오..

text = ''''
저의 홈페이지 주소는 https://www.homepage.com입니다.
당신의 홈페이지 주소는 http://www.yourhomepage.com입니다'''

reg = re.compile(r'http[s]?://w{3}\.[a-zA-Z0-9_-]+\.[a-z]+')
reg.sub('*****', text)
```
