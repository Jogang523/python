```python
def my_ipaddress():
  url = 'https://ip.pe.kr/'
  res = requests.get(url)
  path = "body > div.cover-container.d-flex.p-3.flex-column > div.inner.cover.text-center > h1"
  soup = BeautifulSoup(res.text, 'html.parser')
  ipaddress = soup.select(path)[0].text
  return ipaddress
```

```python
my_ipaddress()
```

```python
def stock_price(code="005930"):
  url = "https://finance.naver.com/item/main.naver?code={}".format(code)
  path = '#chart_area > div.rate_info > div > p.no_today'

  r = requests.get(url)
  soup = BeautifulSoup(r.text, 'html.parser')
  return soup.select(path)[0].text
```

```python
stock_price("005380")
```

```python
stock_price()
```

```python
# *b  : 가변매개변수 - 값을 안 받을수도 있고, 여러개의 값을 받을수도 있음...
```

```python
def sum_f(a, *b):
  print("a:", a)
  print("b:", b)  # 함수 내에서 tuple로 저장
  # return a + b -- error :  정수 + tuple
  return a + sum(b)
```

```python
sum_f(2, 3,4,5,6)
```

```python
sum_f(2)
```

```python
# 키워드 매개변수 : **b
# 값을 받을 때, key=value형식으로 값을 받는다
# 함수 내부에서는 받은 값을 dictionary로 저장
```

```python
def foo(x, y, **z):
  print("x:", x)
  print("y:", y)
  print("z:", z)
```

```python
foo(2, 4, a=1, b=2, c=3)
```

```python
# 일반매개변수 x, y, 가변매개변수 *c, 키워드 매개변수 **d를 가진 함수에서 입력받은 모든 값의 평균을 리턴하는 함수를 생성하고
# 위의 함수를 실행하시오..

def cal_avg(x, y, *c, **d):
  print(x)
  print(y)
  print(c)
  print(d)
  suma = x + y + sum(c) + sum(d.values())
  avg = suma/ (2+len(c)+len(d))
  return avg
```

```python
cal_avg(2,4, 3,4,5,3, a=2, b=7, c=3)
```
