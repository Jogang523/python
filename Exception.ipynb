```python
# Syntax error - 문법 에러
# Name error - 변수 이름 에러
# Indentation error - 들여쓰기 에러
# Index Error - 인덱싱, 슬라이싱 에러
# Type Error - 데이터타입 에러
# Zero DivisionError - 0으로 나눌 떄
```

```python
a = 5
b = 0

try:
  c = a/b
  print(c)
except ZeroDivisionError as e:
  print(e)
  print("0으로 나눌 수 없습니다.")
```
