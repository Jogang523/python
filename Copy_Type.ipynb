```python
# 단순 복사
# 얕은 복사
# 깊은 복사
```

```python
a = [1,2,3]

# 메모리의 어느 물리적 공간이 10생성.
# a = 10의 메모리 주소
```

```python
# 단순 복사 ('='를 통한 복사)

# 실제로는 값을 복사하는 것이 아니라 주소값을 복사해서 공유...
# 동일한 값을 a와 b가 같이 가리키고 있다..

b = a
print(b)

b[0] = 100  # 원본을 공유하고 있고, 원본 수정..

print('a :', a)
print('b :', b)
```

```python
# 얕은 복사

# 복사해서 새로운 객체를 생성, 원래의 값을 저장한다.. 동일한 객체를 공유하지 않는다..

import copy

a = [1,2, [3,4]]
b = copy.copy(a)  # a를 복사해서 새로운 b를 만든다..

print('a :', a)
print('b :', b)

b[1] = 100

print('a :', a)
print('b :', b)  # b에만 수정된 값이 저장됨... a는 수정이 되지 않음..
```

```python
b[2][0] = 100

print('a :', a)
print('b :', b)  # b를 수정하는데, a에도 반영이 된다..

# 얕은 복사
# 최상위 레벨만 새로운 객체로 복사하고, 내부 객체는 여전히 단순 복사로 주소값을 공유..
```

```python
# 깊은 복사
# 최상위 계층과 그 내부의 모든 객체를 완전히 새로운 객체로 복사해서 생성한다..

a = [1,2, [3,4]]
b = copy.deepcopy(a)  # 깊은 복사


b[1] = 100
b[2][0] = 100

print('a :', a)
print('b :', b)  # 모든 레벨에서 a와 다른 b가 복사되어 생성된다..
```
