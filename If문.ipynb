```python
# Type 2

# 성적을 입력 받은 후 70점 이상이면 "축하합니다"
# 미만이면 "노력하세요"를 출력

score = int(input('점수를 입력하세요'))

if score >= 70:
  print('축하합니다')
else:
  print('노력하세요')
```

```python
# Type 3

# 성적을 입력 받은 후 90점 이상이면 "A학점입니다"
# 성적을 입력 받은 후 80점 이상이면 "B학점입니다"
# 성적을 입력 받은 후 70점 이상이면 "C학점입니다"
# 성적을 입력 받은 후 70점 미만이면 "D학점입니다"
```

```python
score = int(input('점수를 입력하세요'))

if score >= 90:
  print('A학점입니다')
elif score >= 80:
  print('B학점입니다')
elif score >= 70:
  print('C학점입니다')
else:
  print('D학점입니다..')
```

```python
pw = 7878

# 패스워드를 입력 받은 후 패스워드가 일치하면 '환영합니다', 일치하지 않으면
# '패스워드가 틀립니다'가 출력되도록 하시오.

password = int(input("패스워드를 입력하세요"))

if password == pw:
  print('환영합니다')
else:
  print('패스워드가 틀립니다')
```

```python
# 임의의 정수를 입력받은 후 짝수이면 "작수입니다"를 출력, 홀수이면 "홀수입니다"를 출력
num = int(input('임의의 정수를 입력하세요'))

if num%2 == 0:
  print("짝수입니다")
else:
  print("홀수입니다")
```

```python
# 임의의 두 정수 a, b를 입력받은 후
# a가 b보다 크면, "a가 b보다 큽니다"
# a가 b보다 작으면, "a가 b보다 작습니다"
# a와 b가 같으면, "a와 b가 같습니다"를 출력하도록 하시오

a = int(input())
b = int(input())

if a > b:
    print("a가 b보다 큽니다")
elif a < b:
    print("a가 b보다 작습니다")
else:
    print("a와 b가 같습니다")
```

```python
# 출생년도를 입력받은 후, (age = 현재년도 - 출생년도)
# 60세 이상이면 "노년이시군요"
# 40세 이상 60세 미만이면 "중년이신군요"
# 20세 이상 40세 미만이면 "청년이시군요"
# 20세 미만이면 "유년이시군요"를 출력

import datetime
now = datetime.datetime.now()
current_year = now.year

birth_year = int(input('출생년도를 입력하세요'))
age = current_year - birth_year

if age >= 60:
    print('노년이시군요')
elif age >= 40:
    print('중년이신군요')
elif age >= 20:
    print('청년이시군요')
else:
    print('유년이시군요')
```

```python
names = {'Mary':1024, 'Tom':5412, 'Bob':2145}

# 이름을 입력받은 후 이름이 names에 있으면 값을 출력하고, 없으면 값 1000으로
# names에 추가하시오 단, setdefault를 사용하지 마시오

name = input('이름을 입력하세요')

# names.setdefault(name, 1000)

if name in names:
  print(names[name])
else:
  names[name] = 1000
  print(names)
```
