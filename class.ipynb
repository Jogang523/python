```python
# 원의 면적과 원둘레를 계산하는 클래스를 생성..

# 변수(속성) : 반지름 - 인스턴스 변수
# 메소드 : 원면적계산, 원둘레계산
# pi : 클래스 변수

class Circle:

  pi = 3.14  # 클래스 변수 - 모든 인스턴스에서 동일한 값을 가지는 변수

  def __init__(self, radius):  # 생성자함수 : 인스턴스 생성 시 자동 실행..
    self.radius = radius   # 일반매개변수로 값을 받아서 인스턴스변수에 할당..

  def area(self):
    return Circle.pi * (self.radius**2)  # 클래스변수 호출할 경우 클래스명.클래스변수

  def circumferences(self):
    return 2 * Circle.pi * self.radius
```

```python
cir = Circle()
```

```python
cir = Circle(3)
```

```python
cir.area()
```

```python
cir.circumferences()
```

```python
cir1 = Circle(5)
```

```python
cir1.area()
```

```python
cir1.circumferences()
```

```python
# 클래스 : FourCal

# 인스턴스 변수 a, b - 정수타입
# 메소드 : addf, minusf, multf, divf
# 생성자 메소드를 포함

# 위의 클래스를 만들고, 인스턴스를 생성해서 각각의 메소드를 호출하시오..
```

```python
class FourCal:

  def __init__(self, a ,b):
    self.a = a
    self.b = b

  def addf(self):
    return self.a + self.b

  def minusf(self):
    return self.a - self.b

  def multif(self):
    return self.a * self.b

  def divf(self):
    return self.a / self.b
```

```python
fc = FourCal()
```

```python
fc = FourCal(2,4)
```

```python
fc.addf()
```

```python
fc.minusf()
```

```python
fc.multif()
```

```python
import math

class Circle:
  def __init__(self, radius):
    self.__radius = radius  # 인스턴스 변수 & 은닉

  def get_area(self):
    return math.pi * self.__radius**2

  def get_perimeter(self):
    return math.pi * self.__radius * 2

  @property              # getter를 만들어주는 decorator => getter : 은닉된 변수값을 출력
  def get_radius(self):
    return self.__radius

  @get_radius.setter     # setter를 만들어주는 decorator => setter : 은닉된 변수의 값을 수정..
  def set_radius(self, radius):
    self.__radius = radius
```

```python
cir  = Circle(5)
```

```python
cir.get_area()
```

```python
cir.get_perimeter()
```

```python
cir.__radius
```

```python
cir.set_radius = 10  # setter 함수를 속성처럼 사용...
```

```python

class BankAccount:

  def __init__(self, balance):
    self.__balance = balance

  def deposit(self, amount):
    if amount <= 0:
      raise ValueError('입금액은 0원 이상이어야 합니다.')
    self.__balance += amount

  def withdraw(self, amount):
    if amount <= 0:
      raise ValueError('출금액은 0원 이상이어야 합니다.')
    if self.__balance < amount:
      raise ValueError('잔액이 부족합니다.')
    self.__balance -= amount

  @property
  def balance(self):
    return self.__balance

  @balance.setter
  def balance(self, amount):
    if amount < 0:
      raise ValueError('잔액은 0원 이상이어야 합니다.')
    self.__balance = amount
```

```python
account = BankAccount(10000)
```

```python
account.balance
```

```python
account.balance = 20000
```

```python
account.balance
```

```python
account.deposit(5000)
```

```python
account.balance
```

```python
account.withdraw(10000)
```

```python
# 인스턴스 메소드
# 인스턴스와 관련된 작업을 하는 객체
# 인스턴스 변수에 접근 가능
# 클래스 변수에 접근 가능

# 클래스 메소드
# 클래스 차원과 관련된 작업을 하는 객체
# 인스턴스변수에 접근이 불가능.
# 클래스변수에 접근 가능
# 객체 생성없이 모든 인스턴스에서 동일하게 접근이 가능...

# 스태틱 메소드 - 일반 함수와 유사..
# 인스턴스와 클래스와 독립적으로 작동하는 메소드
# 인스턴스변수에 접근이 불가능.
# 클래스변수에 접근 가능
# 인스턴스와 관련된 함수기능
```

```python
class MyClass:

  cls_var = "class_val"

  def __init__(self, val):
    self.instance_var = val

  # 인스턴스 메소드
  def instance_method(self):
    print("self.instance_var", ':', self.instance_var)  # 인스턴스 변수 - self로 접근 가능
    print("MyClass.cls_var", ':', MyClass.cls_var)     # 클래스 변수 - 클래스이름으로 접근

  # 클래스 메소드
  @classmethod
  def class_method(cls):
    print("self.instance_var", ':', self.instance_var)  # 인스턴스 변수 - self로 접근 가능
    print("MyClass.cls_var", ':', MyClass.cls_var)     # 클래스 변수 - 클래스이름으로 접근

  # 스태틱 메소드
  @staticmethod
  def static_method():
    print("self.instance_var", ':', self.instance_var)  # 인스턴스 변수 - self로 접근 가능
    print("MyClass.cls_var", ':', MyClass.cls_var)     # 클래스 변수 - 클래스이름으로 접근
```

```python
mc = MyClass(10)
mc.instance_method()
```

```python
mc.class_method()
```
