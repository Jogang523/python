```python
keys = ['a','b','c','d']
values = ['가', '나', '다', '라']

# for문을 이용해서 keys를 key로 values를 value로 하는 dictionary를 만드시오..
# d = {"a":"가", "b":"나", ... "d":"라"}
```

```python
d = {}
for i in range(len(keys)):
  d[keys[i]] = values[i]
d
```

```python
d1 = {}

for i in d.items():
  # print(i)
  k = i[0]
  v= i[1]
  d1[v] = k

d1
```

```python
words = ['mountains', 'sea', 'sky', 'river', 'sea', 'cloud', 'sky', 'sea', 'mountains']
# 위의 words에서 중복된 단어를 제거한 unque_words list를 만드시오
# 단, set()를 사용하지 마시오

# list(set(words))

L = []

for w in words:
  if w not in L:
    L.append(w)
L
```

```python
words = ['mountains', 'sea', 'sky', 'river', 'sea', 'cloud', 'sky', 'sea', 'mountains']

#  words에 각 단어가 몇 개씩 들어있는지 값을 value로 하는 dict롤 출력하시오..
# {'mountains':2, 'sea':3, 'sky':2, 'river':1, 'cloud':1}

d = {}

for w in words:
  if w in d:
    d[w] += 1
  else:
    d[w] = 1

d
```

```python
# for i in sequence:
#   실행코드
#   실행코드
#   ...

# while 조건식:
#   실행코드
#   실행코드
#   ...

# while문
# 조건이 True이면 반복실행, 조건이 False이면 반복을 중단..
# 조건식이 계속 True이면 무한반복 => 상황이 만족되면 조건식이 False가 되도록 장치만들어 놓는다..
```

```python
# 무한루프

import time

while True:  # 무한실행
  print('Hello', end=' ')
  time.sleep(1)
```

```python
# False조건

import time
a = 0

while a < 10:
  print('Hello', end=' ')
  time.sleep(1)
  a += 1
```

```python
# 반복문
# continue : 아래 부분을 실행하지 않고 반복문의 처음으로 돌아가서 다음 번 반복을 계속
# break : 반복문을 빠져나온다..
```

```python
text = "python is easy and fun."

# 모음만 리스트에 담으시오.. 단, continue를 사용해서 구현하시오..

L = []

for c in text:
  if c not in 'aeiou':
    continue
  L.append(c)

L
```

```python
# range(11)의 연속값 정수에서 반복문을 통해서 3의 배수만 출력하시오.. 단, continue를 사용해서 구현하시오..
```

```python
for i in range(11):
  if i % 3 != 0:
    continue
  print(i)
```

```python
text = "python is easy and fun."
i = 0

while True:
  print(text[i], end='')
  i += 1
  if i == len(text):
    break
```

```python
text = "python is easy and fun."
i = 0

while True:
  if text[i] == ' ':
    break
  print(text[i], end='')
  i += 1
```

```python
# 반복문 while문을 통해서 range(100)에서 continue 와  break를 이용해서 5,6,7,8,9,10만 출력되도록 하시오..
```

```python
for i in range(100):
  if i < 5:
    continue
  if i > 10:
    break
  print(i)
```

```python
i = 0

while True:
  i += 1
  if i < 5:
    continue
  if i > 10:
    break
  print(i)
```

```python
# 점수를 5번 입력받은 후 점수의 평균을 출력하시오..for문, while문 각각 사용...
```

```python
i = 0
total = 0

while i < 5:
  score = int(input('점수를 입력하세요 : '))
  total += score
  i += 1

total/i
```
